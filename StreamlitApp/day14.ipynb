# for Education Level(Ordinal)

```python
data = pd.read_csv('Salary_Data.csv')
data.head()
```

```python
data.dropna(inplace = True)
```

```python
#To rename name of column
data = data.rename(columns={
    'Education Level':'Education'
})
```

```python
data.head()
```

```python
data['Education'].unique()
```

```python
#Inconsistency ie bacherlo ra bachelor degree yesto xa bhane inconsistencey yesto lai milaune ie kasaile bachelor halyo ra kasari bachelor degree halyo data entry garda 
# so aba yesto lai same banauna parxa
```

```python
data['Education']= data['Education'].replace({
    "Bachelor's": 'Bachelors',
    "Bachelor's Degree" : 'Bachelors',
    "Master's": 'Masters',
     "Master's Degree": 'Masters',
    'PhD': 'phD'
})
```

```python
data['Education'].unique()
```

```python
#Manually diney
#.map ra replace le autai kura ho 
#.map le string to integer ma lagne garxa
data['Education']= data['Education'].map({
    'High School':0,
    'Bachelors':1,
    'Masters':2,
    'phD':3
})
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
```

```python
data = sns.load_dataset('iris')
```

```python
sns.get_dataset_names() # seaborn ma already available vako dataset yo use garna milyo
```

```python
data.head()
```

```python
X = data.drop('species', axis = 1)
y= data.species
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y, test_size = 0.2, random_state = 4)
```

```python
#n_neighbor bhaneko k ko value deko ho
#p=2 default linxa yesle ie euclidean distance linxa as a default distance
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(xtrain,ytrain)
```

```python
help(KNeighborsClassifier)
```

```python
yprediction = knn.predict(xtest)
yprediction
```

```python
#Accuarcy
accuracy = accuracy_score(ytest,yprediction)
print(f'Accuracy:{accuracy:.2%}')
```
