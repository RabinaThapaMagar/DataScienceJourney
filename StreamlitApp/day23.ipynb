## PCA

```python
import pandas as pd
```

```python
data.shape
```

```python
data['Class'].unique()
```

```python
df =data.drop('Class',axis =1 )
df.head()
```

```python
from sklearn.decomposition import PCA
```

```python
#PCA le naya column haru banaidinxa ra dimension pani ghatauxa
#yo 13 ota column hamlai handle garna garo vayo sano banauna paryo so 
#malai 5 ota matra columns chahiyo
```

```python
pca = PCA(n_components = 5)# n component vaneko kati ota column banaune ie 13 ota columns haru lai mix garera naya column banauxa ra data pani change vayera auxa
dfn = pca.fit_transform(df)
```

```python
dfn #numpy array ma banera basyo so change it into dataframe
```

```python
dfnew = pd.DataFrame(dfn)
dfnew.head() # 5 ota column ma mathi ko 13 ota nai column ko information xa 13 ota column mix garera naya column diyo
```

```python
#component ie column kati ota pass garne vanera kasari tahha pauene 5 otama  thorai attaune sakyo 13 ota column ko data haru ni ki tyo vanda 6 7 chahana sakthyo ki
ie 5 vaneko optimal no of column ho ki haina hamle deko wala
```

## Kaiser's Rule

```python
#kaiser's rule:see the eigen values and select those who have eigen values more than 1
```

```python
eigenvalues = pca.explained_variance_
eigenvalues # 1 vanda badi vako values lai ley ie 
```

```python
#Shree Plot similar as elbow method
plt.plot(range(13),eigenvalues)
plt.show()
```
