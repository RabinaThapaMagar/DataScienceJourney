```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('wine.data.csv')
df.head()
```

```python
Feature Selection:(vakai feature haru bata subect choose garinxa) hya vako feature haru ma auta matra subset choose garxa or top 5 column haru select garxa 
PCA:dimension reduce garxa tara yehu bato columns lidaina naya column haru banaidinxa 
```

```python
df= df.drop('Class',axis = 1)
df.head()
```

```python
#pca garna vanda agadi data lai scaling garna parxa
```

```python
#scaling
scaler = StandardScaler()
dfs = scaler.fit_transform(df)
```

```python
dfs=pd.DataFrame(dfs,columns = df.columns)
dfs.head()
#aba datascaling pani vayo ra dataframe ma pani ayo data
```

```python
#Applying PCA
#no of columns nadiyera
#df_pca ie pca lagaisakeko data
pca = PCA()
df_pca=pca.fit_transform(dfs)
df_pca

#numpy aray ma xa so dataframe ma banaune
```

```python
#pca lagaiskae paxi naya columns haru auxa
df_pca = pd.DataFrame(df_pca)
df_pca.head()
#pca vaisakera ako first component lai pca component 1 vanxa ra mathi column ko most of the data ko informraton principle cm
#component 1 ma hunxa
```

```python
#no of component 5 dida sabai column ko data 5 ota ma auxa kinai
```

```python
#kaiser's Rule:each column ko eigen value hunxa 
#Scree plot
```

```python
determinet|A -<I
#eigne value and eigen vector
```

```python
#jasko highest eigen value auxa teslai lina parxa vanxa pca le ra pcale
#eignen value lai pca.exlained_variance_ ma rakeko hunxa
```

```python
eigen = pca.explained_variance_
eigen
```

```python
#column wise eignen value ako xa
#kaiser le pheri eigen value 1 vanda thulo vako value matra liney 
```

```python
#kaiser's rule: select those  eigen value more than 1
```

```python
#explained Variance
variance =pca.explained_variance_ratio_
variance
```

```python
#0.36 ie original data ko 36% data chai first column ma raixan, 19% second column, 11% 3rd column le tara olny 36+19+11=66%data matra aune raixa
#3 ota eigen value lida
```

```python
import matplotlib.pyplot as plt
```

```python
#scree Plot: eigen value lai plot garne vanxa 
plt.plot(range(1,14),eigen)
plt.xlabel('Number of components')
plt.ylabel('Eigen values')
plt.show()
```

```python
#sharp turn(thorai angale) ja nira deikhinxa teti ota component liney
#4 ota component liney ie 4 ma elbow baneko xa
```

```python
#or afai dina ni milxa no of components paila nai
pca = PCA(n_components = 4)
df_4 = pca.fit_transform(dfs)
df_4
```

```python
df_4 = pd.DataFrame(df_4)
df_4.head()
```
