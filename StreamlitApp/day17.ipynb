```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('Diabetes.csv')
data.head()
```

```python
X = data[['Glucose', 'Age']]
y = data.Outcome
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y, test_size = 0.2, random_state = 12)
```

```python
dmodel = DecisionTreeClassifier()
dmodel.fit(xtrain,ytrain)
```

```python
#evaluate
yprediction = dmodel.predict(xtest)
yprediction
```

```python
accuracy = accuracy_score(ytest, yprediction)
print(f'Accuracy:{accuracy:.2%}')
```

```python
#training kai data use garera accuracy nikalxau

training data                              testing
model le yo data ma kaam garxa              testing data model le unseen data ho

duita ma kun data ma model ko performance ramro hola ta ??
traiing data ma model ko performance ramro hunxa
```

```python
#training ma model ko performance kasto xa

ypred_train = dmodel.predict(xtrain) #ie training ko data pass gardiyeu ie jun data bata model train vako tei data bata prediction nikaldai xam
```
