```python
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from mpl_toolkits.mplot3d import Axes3D
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error,r2_score
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('Salary_Data.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data.isnull().sum()
```

```python
X = data[['Age', 'Years of Experience']]
y = data['Salary']
```

```python
X
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,y,test_size=0.2, random_state = 7)
```

```python
linear_model = LinearRegression()
linear_model.fit(X,y)
```

```python
linear_model.intercept_
```

```python
linear_model.coef_
```

```python
predictions = linear_model.predict(X)
print(predictions)
```

```python
ypred = linear_model.predict(xtest)
ypred
```

```python
mae = mean_absolute_error(ytest, ypred)
print(f'MAE:{mae:.2f}')
```

```python
#root square mean error
rmse = np.sqrt(mae)
print(f'RMSE:{mae:.2f}')
```

```python
#R squared(coefficeint of determination)
r2 = r2_score(ytest,ypred)
print(f'r2_Score:{r2:.2f}')
```

```python
age_range = np.linspace(X['Age'].min(), X['Age'].max(), 50)
exp_range = np.linspace(X['Years of Experience'].min(), X['Years of Experience'].max(), 50)
Age_mesh, Exp_mesh = np.meshgrid(age_range, exp_range)

grid_points = np.c_[Age_mesh.ravel(), Exp_mesh.ravel()]
grid_df = pd.DataFrame(grid_points, columns=['Age', 'Years of Experience'])

salary_pred_plane = linear_model.predict(grid_df).reshape(Age_mesh.shape)

# Diagonal regression line where Age = Years of Experience
line_vals = np.linspace(
    max(X['Age'].min(), X['Years of Experience'].min()), 
    min(X['Age'].max(), X['Years of Experience'].max()), 100
)
line_df = pd.DataFrame({'Age': line_vals, 'Years of Experience': line_vals})
salary_pred_line = linear_model.predict(line_df)

# 3D Plotting
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Surface plot of the predicted salary
ax.plot_surface(Age_mesh, Exp_mesh, salary_pred_plane, alpha=0.5, cmap='viridis')

# Scatter plot of actual data
ax.scatter(X['Age'], X['Years of Experience'], y, color='red', label='Actual Data')

# Line plot: regression along diagonal Age=Experience
ax.plot(line_vals, line_vals, salary_pred_line, color='black', label='Diagonal Regression Line')

# Labels
ax.set_xlabel('Age')
ax.set_ylabel('Years of Experience')
ax.set_zlabel('Predicted Salary')
ax.set_title('3D Regression Plane: Salary vs. Age and Experience')
ax.legend()

plt.show()
```
