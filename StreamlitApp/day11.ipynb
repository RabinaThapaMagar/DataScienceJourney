```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
X = data[['Age']]
y= data.Outcome
```

```python
#logistic regrssion le pani line nai fit garxa tara aba line ma herera predict garne haina 0 ra 1 ko formma predict garne ho line le dine value lai function lagayera 0 ra 1 ko value dinxa
```

```python
plt.scatter(data.Age,data.Outcome)
plt.xlabel('Age')
plt.ylabel('Outcome')
plt.show()
```

```python
#equation
z = b0+b1x paila z ko value nikalxa ra sigmoid vane function use garera value nikalxa , sigmoid ko value line ma pass gardinxa
sigmoid funcion: y = sigmoid(z)1/1+e^z
```

```python
#linear regrssion nai garne ho garna chai tes bata ako y ko value as a z mane ra z ko valeu funtion ma halne ra tyo value lai 0 ra 1 ko form ma dinxa
```

```python
z = b0+b1x
y = sigmoid function(z)
#sigmoid le probability ho deko
#equation of sigmoid: 1/(1+e^-z) 0 ra 1 ko bich ma value dinxa value le deko 0.5 vanda badi xa bhane class 1 predict hunxa ra 0.5 ayo bhane threshold 0 predict hunxa
```

```python
xtrain,xtest, ytrain,ytest = train_test_split(X,y ,test_size = 0.2, random_state = 25)
```

```python
log_model = LogisticRegression()
log_model.fit(xtrain,ytrain)
```

```python
pred_new = pd.DataFrame({
    'Age':[50]
})
```

```python
#prediction
#age = 40

print(log_model.predict(pred_new))
```

```python
print(log_model.predict_proba(pred_new))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
accuracy = accuracy_score(ytest, yprediction) # test data ko accuarcy patta lagako
print(accuracy)
```

```python
print(f'Accuracy:{accuracy:.2%}') # accuracy derai ako ramro
```
