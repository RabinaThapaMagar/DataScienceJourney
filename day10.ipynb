# Multiple Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_absolute_error, r2_score
```

```python
data = pd.read_csv('Salary_Data.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace = True)
```

```python
data.isnull().sum()
```

```python
X = data[['Age']]
y = data.Salary
```

```python
X
```

```python
y
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X,y, test_size = 0.2, random_state = 7)
```

```python
model = LinearRegression()
model.fit(xtrain,  ytrain)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
#line ka banekoxa vanera herna lai  training kai data diye ni hunxa or testing data dida ni hunxa
yprediction ko value chahinxa ra
```

```python
yprediction = model.predict(xtrain)
```

```python
#training ko data ma line plot gareko
plt.scatter(data.Age, data.Salary)
plt.plot(xtrain, yprediction, color = 'red')
plt.ylabel('Salary')
plt.xlabel('Age')
plt.show()
```

```python
yprediction_test = model.predict(xtest)
```

```python
#testing ko data ma line plot gareko
plt.scatter(xtest,ytest)
plt.plot(xtest, yprediction_test, color = 'red')
plt.ylabel('Salary')
plt.xlabel('Age')
plt.show()
```

```python
X = data[['Age','Years of Experience']]
y= data.Salary
```

```python
y
```

```python
X
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X,y, test_size = 0.2, random_state = 9)
```

```python
xtrain
```

```python
ytrain
```

```python
xtest
```

```python
ml_model = LinearRegression()
ml_model.fit(xtrain,ytrain)
```

```python
equation : y + b0+b1*age Bb=*yoe
```

```python
ml_model.intercept_
```

```python
ml_model.coef_
```

```python
#age = 30
#years of experience = 5
pred_salary = ml_model.predict([[30, 5]])
print(pred_salary)
```

```python
#Evalution

yprediction = ml_model.predict(xtest)
yprediction
```

```python
mae = mean_absolute_error(ytest, yprediction)
print('MAE:',mae)
```

```python
r2= r2_score(ytest, yprediction)
print(f'r2_Score:{r2:.2f}')
```
