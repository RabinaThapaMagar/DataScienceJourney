```python
from sklearn.datasets import make_blobs
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.svm import SVC
```

```python
#make blobs use garera data generate garna sakinxa
# x ma diuta features vako data generate garxa blobs le 
x                y
x1     x2        
```

```python
#200 is no of data
#centers means kati ota data chaheko
#cluster_sd = data haru kasari seperate vako xa vanera vanxa 
X, y = make_blobs(n_samples = 200, n_features = 2,centers = 2, cluster_std = 1.7, random_state = 7)
```

```python
help(make_blobs)
```

```python
X
# X numpy array ma xa
```

```python
X_df = pd.DataFrame(X, columns= ['A','B'])
X_df
#x ko correspondig data chai y ma xa
```

```python
plt.scatter(X_df.A, X_df.B,c=y )# aba yo point haru lai xuttauna parxa class ma 0 wala lai auta color ani 1 wala lai arko color
# c= y le color diyo yo data lineary seperable ho
#cluster_sd ko kaam aura class ma data haru kati scatter garera baseko xa
# ie sd kam hunu vayo data derai scatter vako xaina
#sd  derai xa bhane scatter derai vako xa data haru
#mathi cluster_sd = 1.7 deko xam ie badi sd deko vayera data point haru pani badi scatter vayo
plt.xlabel('A')
plt.ylabel('B')
plt.show()
```

```python
# aba yo point haru lai xuttauna parxa class ma
# sd derai deko xam aba data point haru badi scatter vayera arko data point mathi pani ayp ie intersect vayo so aba yeslai linearly seperable data vana mildaina kina bhane yeslai aba line le seperate garna mildaina
```

```python
svm = SVC(kernel = 'linear')
svm.fit(X_df,y) # yesma hamle data lai split gareko xainam so aba split nagarekai data deko
#yesari split navako data ie singai data pass gareko xam so aba evalution garna pani mildaina yp data ma
#kinaki testing ko lagi data nai rakhenam hamle
```
